```python
with open("PanglaoDB_markers_27_Mar_2020.tsv") as file:
    reader = csv.reader(file, delimiter="\t")
    tsvData = list(reader)
    header = tsvData[0]
    data = tsvData[1:]
header
```

```python
species = header.index("species")
ogs = header.index("official gene symbol")
ct = header.index("cell type")
```

```python
VECS_DB = 'PubMed-and-PMC-w2v.bin'
```

```python
ct_to_genes = {}
ct_to_genes_no_space = {}

for i in data:
    if not "Hs" in i[species]:
        continue
    if not i[ct] in ct_to_genes:
        ct_to_genes[i[ct]] = []
    ct_to_genes[i[ct]].append(i[ogs])
    removeSpace = "-".join(i[ct].split())
    if not removeSpace in ct_to_genes_no_space:
        ct_to_genes_no_space[removeSpace] = []
    ct_to_genes_no_space[removeSpace].append("-".join(i[ogs].split()))
ct_to_genes_no_space
    
```

```python
print('Loading model...')
model = KV.load_word2vec_format(VECS_DB, binary=True)
print('end')
```

```python
def wordsInModel(ct_to_genes, model, extraCheck=False):
    # Initialize a set for all the words. First storing
    # all of the cell types
    our_cell_words = set(ct_to_genes.keys())

    our_gene_words = set()
    
    # Add all genes to the set of words
    for genes in ct_to_genes.values():
        our_gene_words.update(genes)
    # Get all words in the model
    
    model_words = set(list(model.index_to_key))
    our_words = our_cell_words.union(our_gene_words)
    record = {}
    cellRecord = {}
    cellTraceOrigin = []
    our_words_in_model = []
    our_cell_words_in_model = []
    count = 0
    if not extraCheck:
        # See which words are in the model using set intersection
        our_words_in_model = our_words & model_words
    else:
        for i in our_words:
            record[i] = []
            if i in our_cell_words:
                cellRecord[i] = []
                variants = [i]
                if i[-1] == 's':
                    remove_s = i[0:len(i)-1]
                    variants.append(i[0:len(i)-1])
                    if "-" in remove_s:
                        temp = remove_s.split("-")
                        j = 0
                        while j < len(temp):
                            temp[j] = temp[j][0].upper() + temp[j][1:]
                            j += 1
                        variants.append("-".join(temp))
                    variants.append(remove_s.upper())
                    variants.append(remove_s.lower())
                if "-" in i:
                    temp = i.split("-")
                    j = 0
                    while j < len(temp):
                        temp[j] = temp[j][0].upper() + temp[j][1:]
                        j += 1
                    variants.append("-".join(temp))
                variants.append(i.upper())
                variants.append(i.lower())
                variants = list(set(variants))
                found = False
                for j in variants:
                    if j in model_words:
                        our_words_in_model.append(j)
                        our_cell_words_in_model.append(j)
                        record[i].append(j)
                        cellRecord[i].append(j)
                        cellTraceOrigin.append(i)
                        found = True
                if found:
                    count += 1
            else:
                if i in model_words:
                    our_words_in_model.append(i)
                    record[i].append(i)
                    count += 1
            
    
    # Sort the words so we have set order
    our_words_in_model = sorted(our_words_in_model)

    if not extraCheck:
        print(len(our_words_in_model) / len(our_words))
        return our_words_in_model
    else: 
        print(count / len(our_words))
        return (our_words_in_model, record, cellRecord, our_cell_words_in_model, cellTraceOrigin)
    
```

```python
def tSNE(x, z, origin, m, p):
    tsne = TSNE(n_components=2, metric=m, perplexity=p)
    principalComponentsTSNE = tsne.fit_transform(x)
    principalDfTSNE = pd.DataFrame(data = principalComponentsTSNE
                 , columns = ['principal component 1', 'principal component 2'], index = z)
    finalDfTSNE = pd.concat([principalDfTSNE, origin], axis = 1)
    groupsTSNE = finalDfTSNE.groupby("origin")
    fig, ax = plt.subplots()
    i = 0
    for name, group in groupsTSNE:
        group.plot(kind="scatter", x = "principal component 1", y = "principal component 2", ax=ax, marker="o", label=name, color=MANY_COLORS[i])
        i += 1
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    figure = plt.gcf()
    figure.set_size_inches(32, 18)
    plt.savefig("img/TSNE." + str(m) + "." + str(p) + ".png")
    # for name, group in groupsTSNE:
    #     plt.plot(group["principal component 1"], group["principal component 2"], marker="o", linestyle="", label=name)
    # figureTSNE = plt.gcf()
    # figureTSNE.set_size_inches(32, 18)
    # plt.savefig("img/TSNE." + str(m) + "." + str(p) + ".png")
```
